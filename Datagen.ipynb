```python
file_path = 'dummy_data++/output.txt'  # Thay 'path/to/your/file.txt' bằng đường dẫn đến tệp tin của bạn

# Mở tệp và đọc nội dung
with open(file_path, 'r') as file:
    lines = file.readlines()  # Đọc từng dòng trong tệp và lưu vào list lines

# Tách các đoạn văn từ từng hàng
paragraphs = []
for line in lines:
    paragraphs.extend(line.split('\n'))  # Tách các đoạn văn dựa trên dấu xuống dòng ("\n")

# Loại bỏ các đoạn văn trống
paragraphs = [paragraph for paragraph in paragraphs if paragraph.strip()]
```

```python
import re
import jaconv
def post_process(text):
    text = ''.join(text.split())
    text = text.replace('…', '...')
    text = re.sub('[・.]{2,}', lambda x: (x.end() - x.start()) * '.', text)
    text = jaconv.h2z(text, ascii=True, digit=True)
    return text
```

```python
from trdg.data_generator import FakeTextDataGenerator
import random
import matplotlib.colors as mcolors
from tqdm import tqdm
import os
save_data = []
def generate_random_color():
    # Danh sách các màu cơ bản (đỏ, đen, xanh) và variation đậm, nhạt của chúng
    base_colors = ['#FF0000', '#000000', '#0000FF']  # Đỏ, đen, xanh
    variations = [0.2, 0.4, 0.6, 0.8]  # Variation: 20%, 40%, 60%, 80%

    # Chọn một màu cơ bản ngẫu nhiên
    base_color = random.choice(base_colors)

    # Tạo biến thể đậm hoặc nhạt
    variation = random.choice(variations)
    rgb_color = mcolors.to_rgb(base_color)
    new_color = mcolors.to_hex([min(1, c * variation) for c in rgb_color])

    return new_color
def get_random_font() :
  folder_path = '/content/dummy_data++/fonts'  # Thay 'đường_dẫn_đến_thư_mục' bằng đường dẫn thực tế đến thư mục của bạn

# Lấy danh sách các tệp tin trong thư mục
  files_list = os.listdir(folder_path)

# Chọn ngẫu nhiên một tệp tin từ danh sách
  random_file = random.choice(files_list)
  return os.path.join(folder_path, random_file)
out_dir = '/content/images'
if not os.path.exists(out_dir) :
  os.makedirs(out_dir)
for i, text in tqdm(enumerate(paragraphs)) :
  rd_font = get_random_font()
  rd_alignment = random.randint(0, 2)
  rd_space_width = random.randint(0, 2)
  rd_character_spacing = random.randint(0, 1)

  # Sử dụng hàm để tạo một biến chứa màu ngẫu nhiên
  text_color = generate_random_color()
  random_elements = tuple(random.randint(0, 5) for _ in range(4))

  # Tạo tuple cuối cùng (5, 5, 5, 5) kết hợp với các phần tử ngẫu nhiên
  rd_margins = (5, 5, 5, 5)[:-len(random_elements)] + random_elements
  rd_fit = bool(random.getrandbits(1))  # Lấy một bit ngẫu nhiên và chuyển đổi thành boolean
  rd_background_type = random.randint(0, 1)
  rd_size = random.randint(16, 64)
  rd_blur = random.randint(0, 2)
  print(text)
  FakeTextDataGenerator.generate(
      index = i,
      text = text,
      font = rd_font,
      out_dir = out_dir,
      size = rd_size,
      extension = 'jpg',
      skewing_angle = 3,
      random_skew = True,
      blur = rd_blur,
      random_blur = True,
      background_type = rd_background_type,
      distorsion_type = 0,
      distorsion_orientation = 0,
      is_handwritten = False,
      width = -1,
      alignment = rd_alignment,
      text_color = text_color,
      orientation = 0,
      space_width = rd_space_width,
      character_spacing = rd_character_spacing,
      margins = rd_margins,
      fit = rd_fit,
      output_mask = False,
      word_split =  False,
      image_dir = '',
      name_format = 2,
      )
  save_data.append({
        'text': post_process(text),
        'image_name': f'{i}.jpg'  # Assuming you use index 'i' for image filenames
    })
```

```python
import pandas as pd
df = pd.DataFrame(save_data)
# Saving DataFrame to a CSV file
csv_file_path = '/content/dummy_data_no_duplicate.csv'  # Specify the path where you want to save the CSV file

try:
    df.to_csv(csv_file_path, index=False, encoding='utf-8')
    print(f"CSV file saved at {csv_file_path}")
except Exception as e:
    print("Error occurred while writing to CSV:", e)
```

```python
import zipfile
import os

def zip_folder(folder_path, zip_path):
    # Mở tệp zip cho việc ghi
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # Duyệt qua tất cả các tệp tin và thư mục trong thư mục đích
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                # Tạo đường dẫn đầy đủ cho tệp tin nguồn
                source_file = os.path.join(root, file)
                # Tạo đường dẫn tương ứng trong tệp zip
                target_file = os.path.relpath(source_file, folder_path)
                # Thêm tệp tin vào tệp zip
                zipf.write(source_file, target_file)
```

```python
MAXLEN = 40
import numpy as np
def encoder_data(text, maxlen=MAXLEN):
        text = "\x00" + text
        x = np.zeros((maxlen, len(alphabet)))
        for i, c in enumerate(text[:maxlen]):
            x[i, alphabet.index(c)] = 1
        if i < maxlen - 1:
          for j in range(i+1, maxlen):
            x[j, 0] = 1
        return x
```
